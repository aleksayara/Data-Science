```python
import pandas as pd
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
# отберем только цветки 'versicolor' и 'verginica'
X = iris.data[50:]
Y = iris.target[50:]
```

```python
# сделаем перекодировку значений с 1 и 2 на 0 и 1
for i in range(0,len(Y)):
    if Y[i] == 1:
        Y[i] = 0
print(Y)
```

```python
for i in range(0,len(Y)):
    if Y[i] == 2:
        Y[i] = 1
print(Y)
```

```python
len(Y)
```

```python
# определим начальные коэффициенты регрессии
tetta = np.random.normal(size=(5,))
tetta
```

```python
# Определим матрицу X, добавив в нее столбец из единиц, для умножения на тетта 0
X = np.c_[X, np.ones(X.shape[0])] 
X
```

```python
def lin_reg(x1, x2, x3, x4, x5, t1, t2, t3, t4, t5 ):
    predict = x1*t1 + x2*t2 + x3*t3 + x4*t4 + x5*t5  
    p = 1. / (1 + np.exp(-predict))
    return p
```

```python
P = lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])  
P
```

```python
def cross_entropy(Y, S):

    E = 0

    for i in range(len(P)):
        if Y[i] ==1:
            E -= np.log(S[i])
        else:
            E -= np.log(1 - S[i])
    return E
```

### Градиентный спуск

```python
K = []
for i in range(0,len(Y)):
    K.append(Y[i])
print(K)
```

```python
Y = K
```

```python
Gradloss = []
learning_rate = 0.001
# выбираем начальное приближение
tetta = [-0.1960596 , -0.30755035,  0.38426438,  0.68284465, -0.41863886]
for i in range(10000):
    if i % 10 ==0:
        f = lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])
        tetta -= learning_rate * np.dot((f-Y).T, X)
        Gradloss.append(cross_entropy(Y, f))
        print(cross_entropy(Y, f))
        print(tetta)
```

```python
# Рассчитываем вероятность отнесения объекта к классу 'virginica', так как его кодировали как 1
Y_pred = np.round(lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4]))
Y_pred
```

```python
# считаем точность модели
import sklearn
sklearn.metrics.accuracy_score(Y, Y_pred)
```

### Nesterov Momentum

```python
Nestloss = []
gamma = 0.975
learning_rate = 0.001
v = np.array([0 for i in range(5)])
tetta_n = np.array([-1.73807571, -0.79552272, -0.23703213,  0.91863707,  0.64974838])
for i in range(10000):
    if i % 10 ==0:
        f_n = lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], tetta_n[0], tetta_n[1], tetta_n[2], tetta_n[3], tetta_n[4])
        tetta_n -= (gamma*v + learning_rate * np.sum(np.dot((f_n-Y).T, X)) / len(f_n))
        tetta_n = tetta_n - v
        Nestloss.append(cross_entropy(Y, f_n))
        print(cross_entropy(Y, f_n))
        print(tetta_n)
```

```python
Y_pred_Nes = np.round(lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], \
                              tetta_n[0], tetta_n[1], tetta_n[2], tetta_n[3], tetta_n[4]))
```

```python
sklearn.metrics.accuracy_score(Y, Y_pred_Nes)
```

### RMSProp

```python
RMSloss = []
gamma = 0.9
learning_rate = 0.01
eps = 0.000001
v = np.array([0 for i in range(5)])
tetta_rms = np.array([-1.73807571, -0.79552272, -0.23703213,  0.91863707,  0.64974838])
for i in range(10000):
    if i % 10 ==0:
        f = lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], tetta_rms[0], tetta_rms[1], tetta_rms[2], tetta_rms[3], tetta_rms[4])
        grad =  np.sum(np.dot((f-Y).T, X)) / len(f)
        v = gamma*v + (1-gamma)*grad**2
        tetta_rms = tetta_rms - learning_rate * grad/(np.sqrt(v+eps))
        RMSloss.append(cross_entropy(Y, f))
        print(cross_entropy(Y, f))
        print(tetta_rms)
```

```python
Y_pred_RMS = np.round(lin_reg(X[:,0], X[:,1], X[:,2], X[:,3], X[:,4],\
                              tetta_rms[0], tetta_rms[1], tetta_rms[2], tetta_rms[3], tetta_rms[4]))
```

```python
sklearn.metrics.accuracy_score(Y, Y_pred_RMS)
```

```python
plt.plot(range(1000), Gradloss, label='Gradient loss' )
plt.plot(range(1000), Nestloss, label='Nesterov momentum loss' )
plt.plot(range(1000), RMSloss, label='RMSProp loss' )
plt.xlabel('epoch', fontsize=15)
plt.ylabel('loss', fontsize=15)
plt.legend()
plt.show()
```
