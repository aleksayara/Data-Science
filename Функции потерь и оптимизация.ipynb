```python
# определим начальные коэффициенты регрессии
tetta = np.random.normal(size=(5,))
tetta
```

```python
# Определим матрицу X, добавив в нее столбец из единиц, для умножения на тетта 0
X = np.c_[X, np.ones(X.shape[0])] 
X
```

```python
def lin_reg(x1, x2, x3, x4, x5, t1, t2, t3, t4, t5 ):
    predict = x1*t1 + x2*t2 + x3*t3 + x4*t4 + x5*t5  
    p = 1. / (1 + np.exp(-predict))
    return p
```

```python
P = lin_reg(X[:,4], X[:,0], X[:,1], X[:,2], X[:,3], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])  
P
```

```python
def cross_entropy(Y, S):

    E = 0

    for i in range(len(P)):
        if Y[i] ==1:
            E -= np.log(S[i])
        else:
            E -= np.log(1 - S[i])
    return E
```

### Градиентный спуск

```python
K = []
for i in range(0,len(Y)):
    K.append(Y[i])
print(K)
```

```python
Y = K
```

```python
learning_rate = 0.0001
# выбираем начальное приближение
tetta = [-0.1960596 , -0.30755035,  0.38426438,  0.68284465, -0.41863886]
for i in range(100):
    if i % 10 ==0:
        f = lin_reg(X[:,4], X[:,0], X[:,1], X[:,2], X[:,3], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])
        tetta[0] -= learning_rate * np.sum((f - Y) * X[:, 4])/len(f)
        tetta[1] -= learning_rate * np.sum((f - Y) * X[:, 0])/len(f)
        tetta[2] -= learning_rate * np.sum((f - Y) * X[:, 1])/len(f)
        tetta[3] -= learning_rate * np.sum((f - Y) * X[:, 2])/len(f)
        tetta[4] -= learning_rate * np.sum((f - Y) * X[:, 3])/len(f)
        print(cross_entropy(Y, f))
        print(tetta)
```

### Nesterov Momentum

```python
gamma = 0.9
learning_rate = 0.0001
n = 0.05
v = np.array([0 for i in range(5)])
tetta = np.array([-0.1960596 , -0.30755035,  0.38426438,  0.68284465, -0.41863886])
for i in range(100):
    if i % 10 ==0:
        f = lin_reg(X[:,4], X[:,0], X[:,1], X[:,2], X[:,3], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])
        tetta[0] -= (gamma*v[0] + learning_rate * np.sum((f - Y) * X[:, 4]))/len(f)
        tetta[1] -= (gamma*v[1] + learning_rate * np.sum((f - Y) * X[:, 0]))/len(f)
        tetta[2] -= (gamma*v[0] + learning_rate * np.sum((f - Y) * X[:, 1]))/len(f)
        tetta[3] -= (gamma*v[0] + learning_rate * np.sum((f - Y) * X[:, 2]))/len(f)
        tetta[4] -= (gamma*v[0] + learning_rate * np.sum((f - Y) * X[:, 3]))/len(f)
        print(tetta, cross_entropy(Y, f))
```

### RMSProp

```python
gamma = 0.9
learning_rate = 0.0001
eps = 0.00000001
v = np.array([0 for i in range(5)])
tetta = np.array([-0.1960596 , -0.30755035,  0.38426438,  0.68284465, -0.41863886])
for i in range(100):
    f = lin_reg(X[:,4], X[:,0], X[:,1], X[:,2], X[:,3], tetta[0], tetta[1], tetta[2], tetta[3], tetta[4])
    tetta[0] -= np.sum((f - Y) * X[:, 4])/len(f)
    tetta[1] -= np.sum((f - Y) * X[:, 0])/len(f)
    tetta[2] -= np.sum((f - Y) * X[:, 1])/len(f)
    tetta[3] -= np.sum((f - Y) * X[:, 2])/len(f)
    tetta[4] -= np.sum((f - Y) * X[:, 3])/len(f)
    v = gamma*v + (1-gamma)*tetta**2
    tetta = tetta - learning_rate * tetta/(np.sqrt(v+eps))
    print(tetta, cross_entropy(Y, f))
```
